# **Comparitive Analysis**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5
ind = np.arange(N) 
width = 0.25
```

```python
KmeansVals = [27.67, 25.67, 19.87, 17.54, 9.25]
bar1 = plt.bar(ind, KmeansVals, width, color = 'crimson')
```

```python
MiniBatchKmeansVals = [27.45, 25.53, 19.07, 18.43, 9.52]
bar2 = plt.bar(ind, MiniBatchKmeansVals, width, color='lightgreen')
```

```python
AgglomerativeVals = [38.1, 22.3, 17.8, 12.5, 9.34]
bar3 = plt.bar(ind,AgglomerativeVals, width, color = 'lightskyblue')
```

```python

KmeansVals = [27.67, 25.67, 19.87, 17.54, 9.25]
bar1 = plt.bar(ind, KmeansVals, width, color = 'crimson')

MiniBatchKmeansVals = [27.45, 25.53, 19.07, 18.43, 9.52]
bar2 = plt.bar(ind+width, MiniBatchKmeansVals, width, color='lightgreen')

AgglomerativeVals = [38.1, 22.3, 17.8, 12.5, 9.34]
bar3 = plt.bar(ind+width*2,AgglomerativeVals, width, color = 'lightskyblue')

plt.xlabel("Clustering Algorithms")
plt.ylabel('Percentages of Colors')
plt.title("Comparitive Analysis")
  
plt.xticks(ind+width,['Color-1', 'Color-2', 'Color-3', 'Color-4', 'Color-5'])
plt.legend( (bar1, bar2, bar3), ('Kmeans', 'MiniBatchKmeans', 'Agglomerative') )
plt.show()
```
